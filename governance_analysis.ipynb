Removing “Governance Estimate (-2.5 to +2.5)”
->Governance estimates are measured on a different scale (-2.5 to +2.5), making them less interpretable and not directly comparable with normalized scores

Removing “Number of Sources”
->The number of sources reflects data reliability rather than governance performance, so it was excluded from analytical comparisons.

To ensure consistency and meaningful comparisons,filtered the dataset to include only governance scores (0–100), as other types either used different scales or represented metadata rather than actual performance indicators.

Changes in df_final will NOT affect df_melted

To ensure that the filtered dataset is independent and to avoid unintended side effects or chained assignment issues during further transformations.

```python
df_final=df_melted[df_melted['Type'].str.contains("score", case=False)].copy()
```

6. stability analysis
which countries are stable vs volatile

```python
stability=df_final.groupby('Country Name')['Value'].std().sort_values()
stable_countries=stability.head(7)
unstable_countries=stability.tail(7)
```

high std -> value fluctuate a lot so unstable

```python
country_stats = df_final.groupby("Country Name")["Value"].agg(["mean", "std"]).sort_values("mean", ascending=False)
country_stats.head(14)
```
